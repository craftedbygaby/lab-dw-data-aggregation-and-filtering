1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
low_tca = (marketing['Total Claim Amount'] < 1000) & (marketing['Response'] == 'Yes')
low_tca_df = marketing[low_tca]
low_tca_df
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
marketing.columns
```

```python
# pivot table with Monthly Premium Auto by Policy Type and Gender
responded_yes = marketing[marketing['Response'] == 'Yes']

premium = responded_yes.pivot_table(index='Policy Type', columns='Gender', values=['Monthly Premium Auto', 'Customer Lifetime Value'], aggfunc='mean').round(2)

premium 
```

```python
# compare these insights to total_claim_amount patterns, and discuss which segments appear most profitable or low-risk for the company.
# pivot table of total_claim_amount, gender and policy type

tca = responded_yes.pivot_table(index='Policy Type', columns='Gender', values='Total Claim Amount', aggfunc='mean').round(2)

tca
```

The premiums and the total claims are very similar, but the personal auto average for woman have a 8 moneys difference from the one for man (and it is the highest premium).  The presonal auto also has the highest CLV, so marketing could focus on personal auto policies focused on women

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
customers_per_state = marketing.groupby('State')['Customer'].count()
customers_per_state[customers_per_state > 500]
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_edu_gender = marketing.pivot_table(index='Education', columns='Gender', values='Customer Lifetime Value', aggfunc=['max', 'min', 'median', 'sum']).round(2)

clv_edu_gender
```

```python
# Redid the pivot table without gender as differences between genders are minimum.
clv_edu = marketing.pivot_table(index='Education', values='Customer Lifetime Value', aggfunc=['max', 'min', 'median', 'sum']).round(2)

clv_edu
```

The doctors are the one with the least overall sum, but the range of their policies are the smaller than other eductations levels, while highschool has the highest - might be interesting for sales people to know that they should invest more time in convincing that demographics as their policies seem to be worth more than high school demographics (a matter of time spent vs return)
